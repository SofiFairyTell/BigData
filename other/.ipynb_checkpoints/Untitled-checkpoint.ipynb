```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Функция для загрузки PDF-файлов с веб-страницы
def download_pdfs(url, save_folder):
    # Создать папку, если её нет
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # Отправить GET-запрос к указанной URL-ссылке
    response = requests.get(url)

    # Проверить, что запрос был успешным (код 200)
    if response.status_code == 200:
        # Создать объект BeautifulSoup для анализа HTML-страницы
        soup = BeautifulSoup(response.content, 'html.parser')

        # Найти все ссылки на странице
        links = soup.find_all('a')

        # Проход по найденным ссылкам
        for link in links:
            href = link.get('href')
            if href and href.endswith('.pdf'):
                # Сформировать полный URL для файла PDF
                pdf_url = urljoin(url, href)
                
                # Определить имя файла (последний компонент URL)
                pdf_filename = os.path.join(save_folder, os.path.basename(href))
                
                # Скачать файл PDF и сохранить его на диск
                pdf_response = requests.get(pdf_url)
                with open(pdf_filename, 'wb') as pdf_file:
                    pdf_file.write(pdf_response.content)
                print(f"Скачан файл: {pdf_filename}")

    else:
        print(f"Не удалось получить доступ к странице: {url}")

# Задать URL-ссылку и папку для сохранения файлов
url = "https://the-eye.eu/public/Books/rpg.rem.uz/7th%20Sea%20Guides/01%20For%20The%20Player%20And%20GM/"
save_folder = "G:\\Private\\test"  # Укажите путь к папке на вашем устройстве

# Вызвать функцию для загрузки PDF-файлов
download_pdfs(url, save_folder)
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import tkinter as tk
from tkinter import filedialog

# Функция для загрузки PDF-файлов с веб-страницы
def download_pdfs(url):
    # Отправить GET-запрос к указанной URL-ссылке
    response = requests.get(url, timeout=30)  # Установите таймаут в 30 секунд
    # Проверить, что запрос был успешным (код 200)
    if response.status_code == 200:
        # Создать объект BeautifulSoup для анализа HTML-страницы
        soup = BeautifulSoup(response.content, 'html.parser')

        # Найти все ссылки на странице
        links = soup.find_all('a')

        # Запросить у пользователя выбор папки для сохранения файлов
        root = tk.Tk()
        root.withdraw()  # Скрыть главное окно приложения
        save_folder = filedialog.askdirectory(title="Выберите папку для сохранения файлов PDF")

        # Проход по найденным ссылкам
        for link in links:
            href = link.get('href')
            if href and href.endswith('.pdf'):
                # Сформировать полный URL для файла PDF
                pdf_url = urljoin(url, href)
                
                # Определить имя файла (последний компонент URL)
                pdf_filename = os.path.join(save_folder, os.path.basename(href))
                
                # Скачать файл PDF и сохранить его на диск
                pdf_response = requests.get(pdf_url)
                with open(pdf_filename, 'wb') as pdf_file:
                    pdf_file.write(pdf_response.content)
                print(f"Скачан файл: {pdf_filename}")

        root.destroy()  # Закрыть окно выбора папки

    else:
        print(f"Не удалось получить доступ к странице: {url}")
```

```python
# Задать URL-ссылку
url = "https://the-eye.eu/public/Books/rpg.rem.uz/7th%20Sea%20Guides/01%20For%20The%20Player%20And%20GM/"

# Вызвать функцию для загрузки PDF-файлов
download_pdfs(url)
```
